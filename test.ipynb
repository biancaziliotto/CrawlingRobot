```python
import mediapy as media
import mujoco
import numpy as np
from src.env import Env
from omegaconf import OmegaConf
```

```python
cfg = OmegaConf.load("cfg/config.yaml")

env = Env(cfg)
print("Environment initialized.")
```

```python
env.mj_data.qpos = [-np.pi / 4, np.pi / 2]
mujoco.mj_kinematics(env.mj_model, env.mj_data)
```

```python
with env.renderer as renderer:
    mujoco.mj_forward(env.mj_model, env.mj_data)
    renderer.update_scene(env.mj_data)
    media.show_image(env.renderer.render())
```
